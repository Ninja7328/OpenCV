```python
import cv2
import numpy as np
```

```python
#Load the Image
img1=cv2.imread('Trial.jpg',0)
img2=cv2.imread('Trial.jpg',0)
```

```python
#Initialize the SIFT detector
sift=cv2.SIFT_create()
```

```python
#Find the keypoints and descriptors with ORB
keypoints1,descriptors1=sift.detectAndCompute(img1,None)
keypoints2,descriptors2=sift.detectAndCompute(img2,None)
```

```python
# BFMatcher with default params
bf=cv2.BFMatcher()
```

```python
#KNN matching
matches=bf.knnMatch(descriptors1,descriptors2,k=2)
```

```python
#Apply ratio test
good_matches=[]
for m, n in matches:
  if m.distance < 0.1 * n.distance:
    good_matches.append(m)
```

```python
#Draw the first 10 matches
matched_img=cv2.drawMatches(img1,keypoints1,img2,keypoints2,good_matches,None,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```

```python
#Display the matches
from google.colab.patches import cv2_imshow
cv2_imshow(matched_img)
```
