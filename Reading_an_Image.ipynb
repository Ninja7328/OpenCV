```python
import cv2
img=cv2.imread("Trial.jpg",cv2.IMREAD_COLOR)
img.shape
```

```python
from google.colab.patches import cv_imshow
cv_imshow(img)
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

Find frequency of pixels in range 0-255

```python
histr=cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(histr)
plt.show()
```
