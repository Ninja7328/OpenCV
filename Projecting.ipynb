```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
src_points=np.array([[51,791],[63,143],[444,211],[426,719]], np.float32)
dest_points=np.array([[1,900],[1,1],[501,1],[501,900]], np.float32)
```

```python
matrix=cv2.getPerspectiveTransform(src_points, dest_points)
```

```python
print(matrix)
```

```python
im=cv2.imread("Trial.jpg")
im.shape
```

```python
im=cv2.resize(im,[800,800])
```
